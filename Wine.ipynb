```python
def ph(ph_val):
    if ph_val > 3.5:
        return'highly acidic'
    else:
        return 'moderately acidic'
```

```python
data['acidic_cat'] = data['pH'].apply(ph)
```

```python
data.head(5)
```

```python
plt.figure(figsize=(9, 4))
sns.countplot(data=data, x='acidic_cat', palette='Blues')
plt.ylabel('Count')
plt.xlabel('Acidic Category')
plt.title('Distribution of Acidic Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['pH'], bins=20, kde=True, color='blue')
plt.xlabel('pH Value')
plt.ylabel('Frequency')
plt.title('Distribution of pH Values in Wine Samples')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=data, x='acidic_cat', y='pH', color='blue')
plt.xlabel('Acidic Category')
plt.ylabel('pH Value')
plt.title('Distribution of pH Values within Acidic Categories')
plt.show()
```

```python
wine = data.drop('acidic_cat' ,axis = 1)
```

```python
wine
```

```python
X = wine.drop(columns=['quality'])
y = wine['quality']
```

```python
#svm model
from sklearn import svm
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,random_state = 34, test_size=0.2)
```

```python
model = svm
```

```python
clf = model.SVC()
```

```python
clf.fit(X_train , y_train)
```

```python
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train , y_train_pred)
```

```python
accuracy_score(y_test , y_test_pred)
```
