Handeling missing values on the basis of categorical and numericl 
--

```python
data.isnull().sum()
```

```python
#This is a machine learning algorithm in which we can impute missing values in replace of mean,median for numerical data

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')  # You can also use 'median', 'most_frequent', etc.
data['reviews_per_month'] = imputer.fit_transform(data[['reviews_per_month']])
```

```python
data.isnull().sum()
```

```python
#missing values for categorical values 
data['name'].fillna('Unknown', inplace=True)
data['host_name'].fillna('Unknown', inplace=True)
data['last_review'].fillna('Unknown', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head(12)
```

Standarized Format
--

```python
data['id'] = data['id'].astype(str)
```

```python
data['id'] = data['id'].str.lower()
```

```python
data.head(2)
```

```python
data = data.sort_values(by='id', ascending=True)
```

```python
data = data.reset_index(drop=True)
```

```python
data
```

```python
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(data['number_of_reviews']))
outliers = data[z_scores > 3]  # Identify outliers with a Z-score greater than 3

Q1 = data['number_of_reviews'].quantile(0.25)
Q3 = data['number_of_reviews'].quantile(0.75)
IQR = Q3 - Q1
outliers = data[(data['number_of_reviews'] < (Q1 - 1.5 * IQR)) | (data['number_of_reviews'] > (Q3 + 1.5 * IQR))]
```

```python
outliers
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=outliers,palette='coolwarm')
plt.xlabel= ('Number of reviews')
plt.ylabel('count')
plt.xticks(rotation=45)
plt.show()
```
